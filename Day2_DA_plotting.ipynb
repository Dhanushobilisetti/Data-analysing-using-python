```python

from matplotlib import pyplot as plt
import numpy as np
over1=np.arange(5,51,5)
inr=np.array([25,51,84,131,116,189,220,250,267,297])
ar=np.array([15,41,131,110,151])
over2=np.arange(5,26,5)
plt.plot(over1,inr,label="India",color='c',marker=">",linewidth=2)
plt.plot(over2,ar,label="Australia",color='b',marker="+",linewidth=2)
plt.legend()
plt.xlabel("Overs")
plt.ylabel("Runs")
plt.grid()
plt.legend()
```
