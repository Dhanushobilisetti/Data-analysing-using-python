```python
#Hypothesis Testing
import numpy as np
from scipy import stats
```

```python
#Take two groups of sample Data
data1=np.array([28,21,26,29,23])
data2=np.array([21,27,25,28,19])
print(data1.mean())
print(data2.mean())
```

```python
t_stat,p_value=stats.ttest_ind(data1,data2,equal_var=False)
```

```python
#print the results
print("T-statistic:",t_stat)
print("p-value:",p_value)
```

```python
alpha=0.05
```

```python
#Interpre the results:
if p_value < alpha:
      print("Reject the Null hypothesis.they are significant")
else:
      print("Nullhypothesis.They are not significant")
```

```python
import pandas as pd
df=pd.read_csv("/content/Vaccine.csv")
df
```

```python
group_A=df['Efficiency']
group_B=df['Rec_Rate']
t_stat,p_value=stats.ttest_ind(group_A,group_B)
print("T-statistic:",t_stat)
print("p-value:",p_value)
```

```python
#Interpre the results:
if p_value < alpha:
      print("H0-related")
else:
      print("H1-Not Related")
```

```python
#ANOVA(Oneway ANOVA)
dfA=df[df['Vaccine']=='A']['Efficiency']
dfB=df[df['Vaccine']=='B']['Efficiency']
dfA
```

```python
f_stat,p_value=stats.f_oneway(dfA,dfB)
print("F-statistic:",f_stat)
print("p-value:",p_value)
```
